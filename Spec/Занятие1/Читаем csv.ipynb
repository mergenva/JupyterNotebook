Задача 1:
Прочитать этот файл.
а) Вывести на экран пробег, год, цену ( в исходном виде ).
б) Вывести пробег в милях, а цену - в долларах ( курс доллара любой ). Намек на то, что надо мили превратить в float.
в) Взять последнюю колонку с двигателем и вывести отдельно мощность, объем и вид топлива.

````python
f = open('Porsche.csv', 'rt', encoding='utf-8-sig') # sig чтобы убрать первый символ
````

````python
title = f.readline() # прочитали одну строку и перенесли каретку на 2 строку
data = f.readlines()
````

````python
data
````

````python
# a)
for line in data:
    cells = line.split(';')
    # print(cells)
    kmage = cells[1]
    price = cells[3]
    year = cells[4]

    print(kmage, price, year)
    
````

````python
def mile(km):
    return km / 1.609

def usd(rub):
    return rub / 80
````

````python
# б)
for line in data:
    cells = line.split(';')
    # print(cells)
    try:
        kmage = round(mile(float(cells[1].replace(' ','').replace('км', ''))), 2)
    except ValueError:
        kmage == 'Новый'
        kmage = 0

    
    price = usd(int(cells[3].removesuffix(' ₽').replace(' ','')))
    year = cells[4]
    print(kmage, price)
````

````python
# в)
for line in data:
    cells = line.split(';')
    full = cells[-1].split('/')
    (objem, sily, fuel) = full
    objem = objem.removesuffix(' л')
    sily = sily.removesuffix('л.с.')    
    print(objem, sily, fuel)
````
