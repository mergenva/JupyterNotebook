Задача 3

1. Прочитать Скатерти.**xlsx**
2. Собрать в список все значения из первой колонки
3. Найти размер и артикул там где они есть, вывести на экран
4. Вывести одтельно длину и ширину каждого товара. Длина - это больший из двух размеров.

```
[0-9]{2,3}[*xх][0-9]{2,3} - размер скатерти
арт\.[0-9\-А-Я]{5,15} - артикул
```

5. Собрать выходные данные в список словарей и сохранить в файл

````python
import re
import openpyxl
````

````python
wb = openpyxl.load_workbook(r"Z:\Python-BI июнь 2023\Датасеты\Скатерти.xlsx")
ws = wb.active

data = []
for cell in ws['A'][2:]:  # пропустили первые две строки
    if cell.value is None:
        break  # данные закончились
    
    match_size = re.search(r'([0-9]{2,3})[*xх]([0-9]{2,3})', cell.value, flags=re.IGNORECASE)
    if match_size is not None:
#         length = int(match_size.group(1))
#         width = int(match_size.group(2))
        # Чтобы длина была больше ширины
        width, length = sorted([int(match_size.group(1)), int(match_size.group(2))])
    else:
        length = None
        width = None
        
    match_art = re.search(r'арт\.(Обр.)?[0-9\-А-Я]{5,15}', cell.value, flags=re.IGNORECASE)
#     if match_art is not None:
#         art = match_art.group(0)
#     else:
#         art = None
    art = match_art.group(0) if match_art is not None else None

    row = {
        "Артикул": art,
        "Длина": length,
        "Ширина": width
    }
    data.append(row)
    print(art, length, width)
````
