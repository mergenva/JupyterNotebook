```python
import xmldict
import json
```

```python
xml_text = """
<storage_domain>
    <name>myisos</name>
    <description>My ISOs</description>
    <type>iso</type>
    <storage>
        <type>nfs</type>
        <address>mynfs.example.com</address>
        <path>/exports/myisos</path>        
    </storage>
    <host>
        <name>myhost</name>        
    </host>
</storage_domain>
"""
```

```python
data = xmldict.xml_to_dict(xml_text)
data
```

```python
json_text = json.dumps(data, indent=4)
print(json_text)
```
