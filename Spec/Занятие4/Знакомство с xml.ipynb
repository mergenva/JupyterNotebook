```python
# https://docs.python.org/3/library/xml.etree.elementtree.html
from xml.etree import ElementTree

# Есть и сторонние библиотеки по работе с XML. Например, xmldict
```

```python
f = open('Валюта.xml', 'rt')
data = ElementTree.parse(f) # можно передавать просто путь к файлу

valcurs = data.getroot()
valcurs
```

```python
valcurs.attrib
```

```python
valcurs.attrib['ID']
```

Дата - это _значение атрибута_ Date _элемента_ Record, вложенного в элемент Valcurs
Цена - это вложенный текст элемента Value, вложенного в элемент Valcurs

```python
for record in valcurs:
    # print(record)
    date = record.attrib['Date'] # строка
    price = record.find('Value').text
    print(date, price)
```
