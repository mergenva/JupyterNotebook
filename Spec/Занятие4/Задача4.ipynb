а) Прочитать Болты.xlsx
б) Найти размеры (диаметр и длина). Размеры могут быть дробные с точкой или запятой,
диаметр может быть вида М6. Если размер только один (диаметр) и он начинается с буквы М,
тоже надо распознать. Одиночные размеры без буквы М можно игнорировать.
в) Где есть количество в штуках или кг - тоже распознать.

г) Наименование, диаметр, длина, количество, вес - сохранить в виде таблицы (в любом виде)

д) Используя вторую таблицу "Размеры болтов" посчитать вес болтов, у которых указано только
количество и наоборот.

```python
import re
import openpyxl
```

```python
regex_size = r'"[MM]?\(?(\d+[\.,]?\d*)\)?[xx*](\d+[\.,]?\d*)|[MM](\d+[\.,]?\d*)"gm'
regex_quantity = '\((\d+) шт\.\)'
regex_weight = '\((\d+) кг\)'
```

```python
wb = openpyxl.load_workbook(r"Болты.xlsx")
ws = wb.active

for cell in ws['A'][1:]:
    print(cell.value)

    match_size = re.search(regex_size, cell.value)
    if match_size is not None:
        if match_size.group(1) is not None:
            diameter = match_size.group(1)

        else:
            # Это гайка (сработала вторая часть выражения после "|")
            diameter = match_size.group(3)
        length = match_size.group(2) # строка или None
    else: # нет размера
        diameter = None
        length = None
        
    match_quantity = re.search(regex_quantity, cell.value)
    quantity = match_quantity.group(1) if match_quantity is not None else None

    match_weight = re.search(regex_weight, cell.value)
    weight = match_weight.group(1) if match_weight is not None else None
    
    print(diameter, length, quantity, weight)
```
