Задача 3

1. Прочитать Скатерти.**xlsx**
2. Собрать в список все значения из первой колонки, начиная с 3 строки
3. Найти размер и артикул там где они есть, вывести на экран
4. Вывести отдельно длину и ширину каждого товара. Длина - это больший из двух размеров.

```
[0-9]{2,3}[*xx][0-9]{2,3} - размер скатерти
арт\.[0-9\-A-Я]{5,15} - артикул
```
5. Собрать выходные данные в список словарей и сохранить в файл

````python
import re # сначала встроенные библиотеки
import openpyxl # затем сторонние библиотеки
````

````python
wb = openpyxl.load_workbook('СкатертиХ.xlsx')
ws = wb.active

data = []
for cell in ws['A'][2:]:
    if cell.value is None:
        break # Данные закончились

    match_size = re.search(r'([0-9]{2,3})[*xx]([0-9]{2,3})', cell.value, flags=re.IGNORECASE)
    if match_size is not None:
        # length = int(match_size.group(1))
        # width = int(match_size.group(2))
        
        # чтобы длина была больше ширины
        length, width = sorted([int(match_size.group(1)), int(match_size.group(2))])

    else:
        length = None
        width = None
        
    match_art = re.search(r'арт\.(Обр.)?[0-9\-A-Я]{5,15}', cell.value, flags=re.IGNORECASE)
    art = match_art.group(0) if match_art is not None else None

    row = {
        "Артикул": art,
        "Длина": length,
        "Ширина": width
    }
    #print(art, length, width)
    data.append(row)
````

````python
import pickle

f = open('скатерти.pickle', 'wb')
# wb - write bites (write binary)
pickle.dump(data, f)
f.close()
````
