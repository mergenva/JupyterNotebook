```python
import requests
import bs4
```

```python
response = requests.get('https://www.specialist.ru/dictionary/definition/python')
response
```

```python
response.text
```

```python
page = bs4.BeautifulSoup(response.text)
```

```python
for element in page.find_all('h3', class_='dp-course-name'):
    print(element['title'])
```

```python
items = page.find_all('li', class_='dp-cards-item')
for item in items:
    print(item.find('h3', class_='dp-course-name')['title'])
    try:
        print(item.find('div', class_='dp-date-text').text.strip())
    except AttributeError:
        print('быз даты')
    print(item.find('div', class_='dp-price-discount').text)
```
